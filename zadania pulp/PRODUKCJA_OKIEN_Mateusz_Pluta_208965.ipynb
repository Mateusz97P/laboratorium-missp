Drugim z kolei zadaniem było rozwiązanie problemu dotyczącego zmaksymalizowania zysków z produkcji okien aluminiowych i drewnianych. Dane początkowe zostały zaproponowane przez prowadzącego i wpisane do poniższych tabel.

![](https://github.com/Mateusz97P/laboratorium-missp/blob/main/zdj%C4%99cia/tabela%20zysk.png?raw=true)

W napisanym programie użyto również suwaków do zmiany wartości zysku z produkcji okien drewnianych oraz aluminiowych.

Stworzony program zwraca nam ilość wyprodukowanych okien oraz zysk z wyprodukowania obu rodzajów okien. 

Dla optymalnych wartości wpisanych do tabeli program zwraca informację o cenie jednej puszki jedzenia oraz o zawartości każdego z produktów w postaci:

Jakość: Optimal

Wytworzono:
Okna_aluminum : 0
Okna_drewno : 4

Zysk: 23372.0 zł

```
from pulp import *
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
import ipywidgets as widgets
import pandas as pd

style = {'description_width': 'initial'}

okna_aluminium_zysk_suw = IntSlider (min=0,max = 10000, value= 3000, description="okna_aluminiowe_zysk", style = style)
okna_drewno_zysk_suw  = IntSlider (min=0,max = 10000, value= 5000, description="okna_drewniane_zysk",  style = style)
fabrykaF1_praca_suw = IntSlider (min=0,max = 20, value= 4, description="fabryka1_czas_pracy", style = style)
fabrykaF2_praca_suw = IntSlider (min=0,max = 20, value= 12, description="fabryka2_czas_pracy", style = style)
fabrykaF3_praca_suw = IntSlider (min=0,max = 20, value= 18, description="fabryka3_czas_pracy", style = style)

def produkcja_okien( 
    zysk_aluminium = 3000,
    zysk_drewno = 5000,
    fabryka1_czas_pracy = 4,
    fabryka2_czas_pracy = 12,
    fabryka3_czas_pracy = 18):

    prob = LpProblem("Problem_produkcji_okien",LpMaximize)

    x1 = LpVariable("Okna_aluminum",0,None,LpInteger)       
    x2 = LpVariable("Okna_drewno",0,None,LpInteger)

    prob += zysk_aluminium*x1 + zysk_drewno*x2, "Zysk"

    prob += 1*x1 <= fabryka1_czas_pracy, "Fabryka1_czas_pracy"
    prob += 2*x2 <= fabryka2_czas_pracy, "Fabryka2_czas_pracy"
    prob += 3*x1 + 2*x2 <= fabryka3_czas_pracy, "Fabryka3_czas_pracy"

    prob.solve()
    print("Jakość:", LpStatus[prob.status])

    print("Wytworzono:")
    for v in prob.variables():
      print(v.name, ":", int(v.varValue))
    print("Zysk:",value(prob.objective),"zł")

    shadows = [{'name':name, 'shadow price': c.pi, "slack": c.slack} for name, c in prob.constraints.items()]
    print(""*10+"Shadow_Prices"+""*10)
    print(pd.DataFrame(shadows))

interact(produkcja_okien,
    zysk_aluminium= okna_aluminium_zysk_suw,
    zysk_drewno = okna_drewno_zysk_suw,
    fabryka1_czas_pracy = fabrykaF1_praca_suw,
    fabryka2_czas_pracy = fabrykaF2_praca_suw,
    fabryka3_czas_pracy = fabrykaF3_praca_suw)
```
